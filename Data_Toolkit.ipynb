**THEORY QUESTIONS**

1. What is NumPy, and why is it widely used in Python?
  - NumPy (Numerical Python) is a core library in Python for numerical computing. It provides a high-performance multidimensional array object and tools for working with these arrays. It is widely used due to its efficiency, broadcasting capabilities, and integration with scientific libraries, making it ideal for mathematical and matrix operations in data analysis and machine learning

2. How does broadcasting work in NumPy?
 -  Broadcasting in NumPy allows operations on arrays of different shapes without the need for manual replication. It automatically expands the smaller array across the larger one to match dimensions, enabling efficient vectorized computations without explicit looping.



3. What is a Pandas DataFrame?
 -  A Pandas DataFrame is a two-dimensional, size-mutable, and heterogeneous data structure with labeled axes (rows and columns). It is used for storing and manipulating structured data, similar to a table in a relational database or an Excel spreadsheet.



4. Explain the use of the groupby() method in Pandas.
 -  The groupby() method in Pandas is used to split data into groups based on one or more keys, apply a function (like sum, mean), and combine the results. It is commonly used in data summarization, aggregation, and transformation operations.



5. Why is Seaborn preferred for statistical visualizations?
 -  Seaborn is preferred for statistical visualizations because it is built on top of Matplotlib and provides high-level, aesthetically pleasing charts. It simplifies the creation of complex plots like heatmaps, boxplots, and pair plots with built-in themes and support for data frames.



6. What are the differences between NumPy arrays and Python lists?
 -  NumPy arrays are more efficient than Python lists for numerical computations as they use less memory and support vectorized operations. Arrays are homogeneous, allowing element-wise operations, while lists are heterogeneous and require loops for computation.

7. What is a heatmap, and when should it be used?
 - A heatmap is a graphical representation of data where values are depicted by color. It is best used to visualize correlation matrices or any data where intensity or frequency across categories needs to be represented, such as gene expression levels or confusion matrices.



8. What does the term “vectorized operation” mean in NumPy?
 - Vectorized operations refer to performing operations on entire arrays without using explicit loops. NumPy executes these operations at low-level C speed, making them faster and more efficient than traditional Python loops.

9. How does Matplotlib differ from Plotly?
 - Matplotlib is a static, traditional plotting library in Python, ideal for publication-ready graphs. Plotly, on the other hand, supports interactive visualizations with features like zooming, tooltips, and exporting, making it more suitable for dashboards and dynamic reports.



10. What is the significance of hierarchical indexing in Pandas?
 - Hierarchical indexing (multi-indexing) in Pandas allows multiple index levels on rows or columns. It provides a way to work with higher-dimensional data in a 2D DataFrame format, enabling more complex data structures and easier data aggregation and selection.



11. What is the role of Seaborn’s pairplot() function?
 - The pairplot() function in Seaborn is used to create a grid of scatter plots for each pair of numerical variables in a dataset. It also displays the distribution of each variable along the diagonal. This function is useful for exploring relationships and trends among multiple variables simultaneously.



12. What is the purpose of the describe() function in Pandas?
 - The describe() function in Pandas generates summary statistics for numerical columns in a DataFrame. It provides metrics such as mean, standard deviation, min, max, and quartile values, helping to quickly understand the distribution and central tendencies in the data.



13. Why is handling missing data important in Pandas?
 - Handling missing data is crucial in Pandas because incomplete data can lead to incorrect analysis or errors during computation. Pandas provides functions like isnull(), dropna(), and fillna() to detect and handle missing values efficiently, ensuring the reliability of data analysis

14. What are the benefits of using Plotly for data visualization?
 - Plotly offers interactive, web-based data visualizations that allow users to zoom, hover, and export charts easily. It supports a wide range of chart types and integrates well with web apps and dashboards, making it ideal for real-time data presentation and exploratory analysis.



15. How does NumPy handle multidimensional arrays?
 - NumPy handles multidimensional arrays using its core object ndarray, which supports arrays of any number of dimensions. These arrays allow efficient storage, slicing, broadcasting, and mathematical operations across multiple axes, making NumPy powerful for scientific and matrix computations.



16. What is the role of Bokeh in data visualization?
 - Bokeh is a Python library for creating interactive and real-time visualizations in web browsers. It enables detailed visual analytics with interactivity such as zooming, panning, and tooltips. Bokeh is suitable for building complex dashboards and integrating visualizations into web applications.



17. Explain the difference between apply() and map() in Pandas.
 - The map() function in Pandas is used to apply a function to each element in a Series, typically for transforming values. The apply() function is more flexible and can be used with both Series and DataFrames, allowing the application of functions across rows or columns.



18. What are some advanced features of NumPy?
 - Advanced features of NumPy include broadcasting, vectorized operations, masking, advanced indexing, linear algebra operations (e.g., matrix multiplication, eigenvalues), and integration with C/C++ code. These features enable efficient scientific computing and data processing.

19. How does Pandas simplify time series analysis?
 - Pandas simplifies time series analysis by providing built-in support for date-time indexing, frequency conversion, resampling, and time-based grouping. Functions like resample() and rolling() allow users to handle trends, seasonality, and date-based slicing effortlessly.



20. What is the role of a pivot table in Pandas?
 - A pivot table in Pandas reshapes and summarizes data by aggregating values based on specified row and column indices. It helps in data analysis by transforming long-format data into a more readable, summarized table for comparison and trend identification.



21. Why is NumPy’s array slicing faster than Python’s list slicing?
 - NumPy’s array slicing is faster than Python’s list slicing because NumPy arrays are stored in contiguous memory blocks and leverage low-level optimizations. Unlike Python lists, which are collections of objects, NumPy arrays are typed and support fast, element-wise operations.



 22. What are some common use cases for Seaborn?
  - Seaborn is commonly used for statistical data visualization. Typical use cases include creating distribution plots (like histograms and KDE plots), categorical plots (like boxplots and violin plots), correlation heatmaps, and pairwise relationships using pairplot(). It is particularly useful when exploring patterns and trends in structured datasets using high-level syntax.

```python
#1. How do you create a 2D NumPy array and calculate the sum of each row python?

import numpy as np

# Create 2D array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Row-wise sum
row_sum = np.sum(arr, axis=1)

print("2D Array:\n", arr)
print("Sum of each row:", row_sum)
```

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame?

import pandas as pd

# Sample DataFrame
data = {'Name': ['A', 'B', 'C'], 'Marks': [80, 90, 85]}
df = pd.DataFrame(data)

# Mean of 'Marks'
mean_value = df['Marks'].mean()
print("Mean of Marks column:", mean_value)
```

```python
#3. Create a scatter plot using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y, color='red')
plt.title('Sample Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```

```python
#4. Calculate the correlation matrix using Seaborn and visualize it with a heatmap

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
df = pd.DataFrame({
    'Math': [90, 85, 80],
    'Science': [88, 86, 82],
    'English': [78, 75, 72]
})

# Correlation matrix
corr = df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#5. Generate a bar plot using Plotly

import plotly.express as px
import pandas as pd

# Data
data = pd.DataFrame({
    'Subjects': ['Math', 'Science', 'English'],
    'Scores': [85, 90, 80]
})

# Bar plot
fig = px.bar(data, x='Subjects', y='Scores', title='Scores by Subject')
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column

import pandas as pd

# Original DataFrame
df = pd.DataFrame({'Name': ['A', 'B', 'C'], 'Marks': [80, 75, 90]})

# Adding new column
df['Grade'] = df['Marks'].apply(lambda x: 'A' if x >= 85 else 'B')

print(df)
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Element-wise multiplication
result = a * b
print("Result of element-wise multiplication:", result)
```

```python
#8. Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 7]

plt.plot(x, y1, label='Line 1', color='blue')
plt.plot(x, y2, label='Line 2', color='green')
plt.title('Multiple Lines Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd

df = pd.DataFrame({
    'Name': ['X', 'Y', 'Z'],
    'Marks': [55, 75, 65]
})

# Filter rows with Marks > 60
filtered_df = df[df['Marks'] > 60]
print(filtered_df)
```

```python
#10. Create a histogram using Seaborn to visualize a distribution

import seaborn as sns
import matplotlib.pyplot as plt

data = [45, 55, 60, 62, 65, 68, 72, 75, 78, 80]

# Histogram
sns.histplot(data, bins=5, kde=True, color='purple')
plt.title("Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
#11. Perform matrix multiplication using NumPy

import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiplication
result = np.dot(A, B)
print("Matrix multiplication result:\n", result)
```

```python
#13. Create a 3D scatter plot using Plotly

import plotly.express as px
import pandas as pd

# Sample data
df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 15, 13, 17],
    'z': [5, 10, 15, 20]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
